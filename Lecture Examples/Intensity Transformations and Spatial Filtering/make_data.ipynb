```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
width = 9*24+17
height = 2*17+210

I = np.zeros([height, width], dtype=np.uint8)
```

```python
for i in range (1,10):
  I[18:18+210-1,i*17+1+(i-1)*7:i*17+1+i*7-1]=255
```

```python
plt.imshow(I)
plt.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
#save the image
im = Image.fromarray(I)
im.save("test_pattern.jpg")
```
